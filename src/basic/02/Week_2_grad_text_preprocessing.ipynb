## 1-1. 단어 및 문장 토큰화 예시

```python
# konlpy에서 제공하는 형태소 분석기: Okt(Open Korea Text) # Komoran(코모란), # Hannanum(한나눔), # 꼬꼬마(Kkma)
from konlpy.tag import Okt
from konlpy.tag import Komoran
from konlpy.tag import Hannanum
from konlpy.tag import Kkma

okt = Okt()
komoran = Komoran()
hannanum = Hannanum()
kkma = Kkma()

text_kor = "안녕하세요 여러분 만나서 반갑습니다ㅋㅋㅋ. 이제 자연언어 클래스를 시작해봅시다:)"

print("Okt: ", okt.morphs(text_kor))
print("Komoran: ", komoran.morphs(text_kor))
print("Hannanum: ", hannanum.morphs(text_kor))
print("Kkma: ", kkma.morphs(text_kor))
```

## **2. 정규 표현식(Regular Expression)**

```python
import re  # 정규 표현식을 위한 모듈

text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

# 주어진 텍스트에서 알파벳을 제외한 모든 문자를 공백으로 처리
preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)

# 처리결과 출력
print(preprocessed_text)
```

```python
text = "I was wondering if anyone out there could enlighten me on this car."

# 주어진 텍스트에서 길이가 1 ~ 2인 단어들을 삭제
shortword = re.compile(r'\W*\b\w{1,2}\b')

# 처리결과 출력
print(shortword.sub('', text))
```

```python
from nltk.tokenize import RegexpTokenizer  # 정규 표현식 토크나이저

text = "Don't be fooled by the name, Mr. Jone's Orphanage is a bright place."

# 문자, 숫자, '(apostrophe)를 포함해 단어 구분
tokenizer = RegexpTokenizer(r"[\w]+")

# 토큰화 결과
print(tokenizer.tokenize(text))
```

## **3. 표제어 추출과 어간 추출**

## 3-1. 표제어 추출

```python
from nltk.stem import WordNetLemmatizer # NLTK의 표제어 추출 도구

lemmatizer = WordNetLemmatizer()

words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']

print('표제어 추출 전 :',words)
print('표제어 추출 후 :',[lemmatizer.lemmatize(word) for word in words])

print('\n')

# 각 단어의 동사 품사로 표제어를 추출
print('dies의 동사 표제어: ', lemmatizer.lemmatize('dies', 'v'))
print('watched의 동사 표제어: ', lemmatizer.lemmatize('watched', 'v'))
print('has의 동사 표제어: ', lemmatizer.lemmatize('has', 'v'))
```

## 3-2. 어간 추출

```python
from nltk.stem import PorterStemmer # 포터 스테머(stemming) 모듈
from nltk.stem import LancasterStemmer  # 랭커스터 스테머(stemming) 모듈

porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()

words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives',
         'fly', 'dies', 'watched', 'has', 'starting']

print('어간 추출 전: ', words)

# 포터 스테머로 어간 추출
print('PorterStemmer 어간 추출 후:', [porter_stemmer.stem(w) for w in words])

# 랭커스터 스테머로 어간 추출
print('LancasterStemmer 어간 추출 후:', [lancaster_stemmer.stem(w) for w in words])
```

## **4. 불용어 제거와 품사 태깅**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words_list = stopwords.words('english')

print('영어 불용어 개수 :', len(stop_words_list))
print('영어 불용어 10개 출력 :',stop_words_list[:10])
```

```python
# 영어 불용어 제거 예시
text = "Family is not an important thing. It's everything."

stop_words = set(stopwords.words('english'))  # 영어 불용어 집합

word_tokens = word_tokenize(text)  # 토큰화 실행

result = []
for word in word_tokens:
    if word not in stop_words:  # 불용어가 아니면 result에 추가
        result.append(word)

print('불용어 제거 전: ', word_tokens)
print('불용어 제거 후: ', result)
```

```python
print(nltk.pos_tag(result))  # 토큰에 품사를 태깅
```

```python
# 한국어 불용어는 일반적으로 사용자가 정의하여 사용
from konlpy.tag import Okt

okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
```

```python
print(okt.pos(example))
```

## **5. 서브워드 토크나이징**

```python
from transformers import AutoTokenizer # LLM에서 주로 활용하는 토크나이저를 사용하기 위한 라이브러리

tokenizer = AutoTokenizer.from_pretrained("gpt2") # 바이트 페어 인코딩(Byte Pair Encoding, BPE)이 적용된 토크나이저, 바이트 기준 토큰화
```

```python
# 영어 서브워드 토크나이저 적용 예시
sentence = "Don't be fooled by the name, Mr. Jone's Orphanage is a bright place."

tokens = tokenizer.tokenize(sentence)

print("Tokens:", tokens)
```

```python
# 한국어 서브워드 토크나이저 적용 예시
text_kor = "안녕하세요, 여러분. 만나서 반갑습니다. 이제 자연언어 클래스를 시작해봅시다!"

tokens = tokenizer.tokenize(text_kor)

print("Tokens:", tokens)
```

```python
# 한국어 서브워드 토크나이저 적용 예시 - 한국어에 특화된 토크나이저
tokenizer_kor = AutoTokenizer.from_pretrained("skt/kogpt2-base-v2")

text_kor = "안녕하세요, 여러분. 만나서 반갑습니다. 이제 자연언어 클래스를 시작해봅시다!"

tokens = tokenizer_kor.tokenize(text_kor)

print("Tokens:", tokens)
```
